## Sun, 10s

```python
#Paramètres utilisés pour la mesure (par défaut)
sampleRate = 2.048e6
centerFreq = 1420e6
centerFreqGHz = centerFreq / 1e9
nfft = 1024

#Lire les données RAW
sun_rawdata = np.fromfile("../data/sun10_2025_02_21_13_14_57_raw.dat", np.complex64)
leftofsun_data = np.fromfile("../data/leftofsun10_2025_02_21_13_50_49_raw.dat", np.complex64)
rightofsun_data = np.fromfile("../data/rightofsun10_2025_02_21_13_52_36_raw.dat", np.complex64)
topofsun_data = np.fromfile("../data/topofsun10_2025_02_21_13_44_24_raw.dat", np.complex64)
bottomofsun_data = np.fromfile("../data/bottomofsun10_2025_02_21_13_45_42_raw.dat", np.complex64)
deg2_bottomofsun_data = np.fromfile("../data/bottomofsun10_-12deg_2025_02_21_13_57_34_raw.dat", np.complex64)
deg7_bottomofsun_data = np.fromfile("../data/bottomofsun10_-17deg_2025_02_21_14_00_19_raw.dat", np.complex64)
deg12_bottomofsundata = np.fromfile("../data/bottomestofsun10_-22deg_2025_02_21_14_18_39_raw.dat", np.complex64)
bottomleftofsun_data = np.fromfile("../data/bottomleftofsun10_-12deg_+10min_2025_02_21_14_12_59_raw.dat", np.complex64)
bottomrightofsun_data = np.fromfile("../data/bottomrightofsun10_-12deg_-10min_2025_02_21_14_14_44_raw.dat", np.complex64)
```

```python
fig = plt.figure()
plt.psd(sun_rawdata, Fs=sampleRate, Fc=centerFreq, NFFT=nfft, scale_by_freq=False, label="Sun")
plt.psd(leftofsun_data, Fs=sampleRate, Fc=centerFreq, NFFT=nfft, scale_by_freq=False, label="$+20$m Ra")
plt.psd(rightofsun_data, Fs=sampleRate, Fc=centerFreq, NFFT=nfft, scale_by_freq=False, label="$-20$m Ra")
plt.psd(topofsun_data, Fs=sampleRate, Fc=centerFreq, NFFT=nfft, scale_by_freq=False, label=r"$+5^{\circ}$ Dec")
plt.psd(bottomofsun_data, Fs=sampleRate, Fc=centerFreq, NFFT=nfft, scale_by_freq=False, label=r"$-5^{\circ}$ Dec")
plt.psd(deg2_bottomofsun_data, Fs=sampleRate, Fc=centerFreq, NFFT=nfft, scale_by_freq=False, label=r"$-2^{\circ}$ Dec")
plt.psd(deg7_bottomofsun_data, Fs=sampleRate, Fc=centerFreq, NFFT=nfft, scale_by_freq=False, label=r"$-7^{\circ}$ Dec")
plt.psd(deg12_bottomofsundata, Fs=sampleRate, Fc=centerFreq, NFFT=nfft, scale_by_freq=False, label=r"$-12^{\circ}$ Dec")
plt.psd(bottomleftofsun_data, Fs=sampleRate, Fc=centerFreq, NFFT=nfft, scale_by_freq=False, label=r"$+10$m Ra, $-2^{\circ}$ Dec")
plt.psd(bottomrightofsun_data, Fs=sampleRate, Fc=centerFreq, NFFT=nfft, scale_by_freq=False, label=r"$-10$m Ra, $-2^{\circ}$ Dec")
plt.ylim(-74,-60)
# plt.axhline(-67)
fig.legend(bbox_to_anchor = (1.5,0.65))
plt.show()
```
